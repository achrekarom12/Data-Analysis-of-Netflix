We will now create two new columns named as 'seasons' and 'runtime'. Seasons column will have no. of seasons for tv shows and zero for movies. Similarely runtime will have duration for movie and for tv shows it will have zero.

```python
df4 = df3.copy()
```

```python
df4.head(3)
```

```python
df4[['number', 'unit']] = df4['duration'].str.split(' ', expand = True)
```

```python
df4.head(4)
```

```python
df4.drop(['duration'], axis=1)
```

```python
df4['number'].dtype
```

```python
df4['number'].astype('int').dtype
```

```python
df4['number'].dtype
```

```python
if df4['number'].astype('int') < 20:
    df4['Seasons'] = df4['number']
else: 
    df4['runtime'] = df4['number']
```

### Task to do: 
Develope a logic that will make a new column 'Seasons' and which will contain no. of seasons for tv shows and null for movies. 
Create a new column called 'runtime' which will have screenplay time for movies and null for tv shows.

```python
df4.info()
```

```python
df5 = df4.drop(['duration'], axis=1)
```

```python
df5_movies = df5[df5['type'] == 'Movie']
df5_movies
```

```python
df5_movies = df5_movies.drop(['unit'], axis=1)
df5_movies
```

```python
list(df5_movies)
```

```python
df5_movies.rename(columns = {'number':'runtime'}, inplace = True)
```

```python
df5_movies.head(10)
```

```python
df5_shows = df5[df5['type'] == 'TV Show']
```

```python
df5_shows.head()
```

```python
df5_shows = df5_shows.drop(['unit'], axis=1)
df5_shows.rename(columns = {'runtime':'seasons'}, inplace = True)
```

```python
df5_shows.head()
```

```python
df_to_be_concat = [df5_movies, df5_shows]
df6 = pd.concat(df_to_be_concat)
df6.head()
```

```python
df6
```

```python
df6 = df6.sort_values(by = 'release_year')
df6.head()
```
